```python
data_agg = n.groupby(['director'])[["release_year", "title"]].aggregate({"release_year":['min','max'], "title": "count"})
data_agg.columns = ['_'.join(col) for col in data_agg.columns]
data_agg=data_agg.reset_index()
data_agg
```

```python
data_agg["yrs_active"] = data_agg["release_year_max"] - data_agg["release_year_min"]
data_agg["movie_per_yr"] = data_agg["title_count"] / data_agg["yrs_active"]
data_agg.sort_values("movie_per_yr", ascending=False)
data_agg#most productive director on basis of active year
```
